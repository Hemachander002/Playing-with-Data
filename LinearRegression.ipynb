```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

### Generating random samples

```python
np.random.seed(42)
x = np.random.rand(100)*10
y = x * 3 + np.random.normal(0,3,100)
df = pd.DataFrame({"X":x,"Y":y})
```

```python
df.head()
```

### Splitting the data for model training and testing

```python
x_train , x_test , y_train , y_test = train_test_split(df[["X"]],df["Y"] , test_size = 0.2 , random_state = 42)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
yhat = model.predict(x_test)
```

### Plotting Time!!

```python
plt.figure(figsize=(8, 5))
sns.kdeplot(y_test, label='Actual', fill=True, color='blue')
sns.kdeplot(yhat, label='Predicted', fill=True, color='red')
plt.xlabel('Target Variable')
plt.ylabel('Density')
plt.title('KDE Plot of Actual vs. Predicted Values')
plt.legend()
plt.show()
```
